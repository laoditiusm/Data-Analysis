NULL VALUE FILL - Replace the Null Values with mean values of the Data

```python
from sklearn.impute import SimpleImputer
import numpy as np

imputer = SimpleImputer(missing_values=np.nan,strategy='mean',fill_value=None)

df['Life expectancy '] = imputer.fit_transform(df[['Life expectancy ']])
df['Adult Mortality'] = imputer.fit_transform(df[['Adult Mortality']])
df['Alcohol'] = imputer.fit_transform(df[['Alcohol']])
df['Hepatitis B'] = imputer.fit_transform(df[['Hepatitis B']])
df[' BMI '] = imputer.fit_transform(df[[' BMI ']])
df['Polio'] = imputer.fit_transform(df[['Polio']])
df['Total expenditure'] = imputer.fit_transform(df[['Total expenditure']])
df['Diphtheria '] = imputer.fit_transform(df[['Diphtheria ']])
df['GDP'] = imputer.fit_transform(df[['GDP']])
df['Population'] = imputer.fit_transform(df[['Population']])
df[' thinness  1-19 years'] = imputer.fit_transform(df[[' thinness  1-19 years']])
df[' thinness 5-9 years'] = imputer.fit_transform(df[[' thinness 5-9 years']])
df['Income composition of resources'] = imputer.fit_transform(df[['Income composition of resources']])
df['Schooling'] = imputer.fit_transform(df[['Schooling']])
```

LIFE EXPECTANCY AND EXPENDITURE
---

This scatter plot provides a visual representation of the correlation between life expectancy (dependent variable) and percentage expenditure (independent variable)

1. General Trend:
- For lower life expectancy values (40–60 years), percentage expenditure remains low and relatively consistent, with most points clustered close to the bottom of the Y-axis.
- At higher life expectancy values (70–80+ years), there is a notable increase in variance for percentage expenditure. 
- Some countries or regions spend significantly more (up to 20,000) to achieve or maintain higher life expectancy levels, while others achieve comparable life expectancy with far less expenditure.


2. Key Observations:
Low expenditure at low life expectancy: 
- Countries or regions with lower life expectancy tend to spend very little, which might reflect poor healthcare infrastructure, limited resources, or economic constraints.

Increasing expenditure for high life expectancy: 
- To achieve life expectancy levels above 75–80 years, a higher percentage of expenditure appears necessary for some regions.

Significant outliers:
- Some data points with very high expenditure (>10,000) do not show correspondingly high life expectancy. 
- These could represent inefficiencies, high costs of healthcare systems, or regions dealing with unique challenges (e.g., aging populations or chronic diseases).
- Conversely, there are points where high life expectancy (~80–90 years) is achieved with low expenditure, likely reflecting efficient healthcare systems, preventative care, or healthier population lifestyles.

4. Hypotheses and Implications:
Low spending and low outcomes: 
- For countries with life expectancy below ~60 years, addressing basic healthcare infrastructure, sanitation, and nutrition may be more impactful than increasing expenditure.

Efficient systems vs. resource-intensive systems: 
- The differences in expenditure for similar life expectancy values indicate the varying efficiency of healthcare systems globally. 
- Some countries achieve high life expectancy with minimal expenditure due to effective public health strategies.

Wealthier regions: 
- Higher expenditures at high life expectancy may reflect wealthier regions investing more in prolonging life even marginally, despite diminishing returns.

This plot underscores the non-linear and asymmetrical relationship between life expectancy and percentage expenditure. While increasing spending helps improve life expectancy up to a point, other social, economic, and healthcare system factors likely influence outcomes significantly, particularly in countries that achieve high life expectancy with relatively low expenditure

```python
import matplotlib.pyplot as plt

# Example data (replace these with your actual dataset)
life_expectancy = df['Life expectancy ']
expenditure = df['percentage expenditure']

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(expenditure,life_expectancy, color='blue', alpha=0.7)

# Add labels and title
plt.title('Life Expectancy vs Percentage Expenditure', fontsize=16)
plt.xlabel('Percentage Expenditure', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)

# Show grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

2. How does Infant and Adult mortality rates affect life expectancy?

- The scatter plot highlights the critical importance of reducing infant and adult mortality to improve life expectancy. It shows a clear disparity between countries with robust healthcare systems and those without.

```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated data
np.random.seed(42)
infant_mortality = df['infant deaths']
adult_mortality = df['Adult Mortality']  # Adult mortality rates (per 1000 adults)
life_expectancy = df['Life expectancy ']

# Scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(infant_mortality, life_expectancy, c=adult_mortality, cmap='viridis', alpha=0.8)
plt.colorbar(scatter, label='Adult Mortality Rate (per 1000)')

# Labels and title
plt.title('Effect of Infant and Adult Mortality on Life Expectancy', fontsize=14)
plt.xlabel('Infant Mortality Rate (per 1000 live births)', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)

# Display the plot
plt.grid(alpha=0.3)
plt.show()
```

Question 2 - Analysis Result
-  QUESTION: How does Infant and Adult mortality rates affect life expectancy?

- Color Gradient Represents Adult Mortality Rate:
Darker points (lower adult mortality rates) are clustered toward the upper left (low infant mortality and high life expectancy).
Lighter points (higher adult mortality rates) are more spread out and tend to occur in regions of lower life expectancy.


- Countries with lower Adult Mortality Rates and lower infant Mortality Rates yield higher life expectancy which indicate better healthcare and living conditions.
- Countries with High Adult Mortality Rates and High Infant Mortality Rates yield lower life expectancy
- As identified below in Note 2.1 and Note 2.2, Haiti is the country most affected by High Adult Mortality Rates and High Infant Mortality Rates yield which results in a 36.3 (yrs) life expectancy while Australia records one of the best life expectancies with 86 yrs. Haiti as a developing country records a 682 Adult Mortality rate, Australia as a developed country records a 69 Adult Mortality Rate.
- With that in mind, there are cases where the Adult Mortality is low yet higher infant mortality rates which results in lower life expectacy - This could indeed be due to dataset-specific anomalies or unique circumstances in certain countries (e.g., issues with maternal care, early childhood diseases). Acknowledge the dataset's limitations in explaining such outliers.

1. Negative Correlation Between Infant Mortality Rate and Life Expectancy:
- As Infant Mortality Rate (x-axis) increases, Life Expectancy (y-axis) decreases.
- This makes sense because higher infant mortality rates often indicate poor healthcare systems and living conditions, which also reduce life expectancy.

2. Clusters of Data:

High Life Expectancy Countries:
- Low infant mortality (<250 per 1000) with darker colors indicating lower adult mortality rates. These represent developed countries.
Low Life Expectancy Region:
- High infant mortality (>250 per 1000), often with higher adult mortality. These likely correspond to developing or underdeveloped countries.

3. Outliers:
- There are some points where infant mortality is very high (e.g., >1000) but life expectancy remains in the mid-range (~50–60 years). These could represent specific regions or countries with unique challenges.

4. Correlation Analysis:

Life Expectancy and Infant Mortality:
- Strong negative correlation: higher infant mortality is consistently associated with lower life expectancy.


Life Expectancy and Adult Mortality:
- Negative correlation: as adult mortality rates increase (lighter colors), life expectancy decreases. However, this is less pronounced than the infant mortality effect.

Combined Effect:
Countries with both high infant and adult mortality rates (lighter colors in the lower right of the plot) exhibit the lowest life expectancy. Conversely, countries with low rates for both metrics tend to have high life expectancy.


THE WHO should focus on developing countries like Haiti and recommend and promote the implemetation of healthcare systems and exposure to Healthy living as those in countries like Australia.
